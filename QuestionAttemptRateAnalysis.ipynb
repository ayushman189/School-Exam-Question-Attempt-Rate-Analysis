```python
import pandas as pd
```

```python
data = {
    'Question': ['Q1','Q2','Q3','Q4','Q5'],
    'Attempted_Students': [45, 38, 50, 30, 25],
    'Total_Students': [50, 50, 50, 50, 50]
}

df = pd.DataFrame(data)
df
```

```python
df['Attempt_Rate (%)'] = (df['Attempted_Students'] / df['Total_Students']) * 100
df

```

```python
df['Attempt_Rate (%)'].mean()
```

```python
def classify(rate):
    if rate >= 75:
        return 'Easy'
    elif rate >= 50:
        return 'Medium'
    else:
        return 'Hard'

df['Difficulty'] = df['Attempt_Rate (%)'].apply(classify)
df
```

```python
import matplotlib.pyplot as plt

plt.bar(df['Question'], df['Attempt_Rate (%)'])
plt.xlabel('Questions')
plt.ylabel('Attempt Rate (%)')
plt.title('Question-wise Attempt Rate')
plt.show()
```
